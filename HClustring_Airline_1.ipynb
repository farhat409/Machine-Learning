```python
import numpy as np
import pandas as pd
import matplotlib.pylab as plt
```

```python
air=pd.read_csv("EastWestAirlines.csv")
```

```python
air
```

```python
def norm_func(i):
    x = (i-i.min())/(i.max()-i.min())
    return (x)
```

```python
df_norm = norm_func(air.iloc[:,1:])
df_norm
```

```python
from sklearn.preprocessing import MinMaxScaler
trans = MinMaxScaler()
data = pd.DataFrame(trans.fit_transform(air.iloc[:,1:]))
data 
```

```python
from scipy.cluster.hierarchy import linkage 
import scipy.cluster.hierarchy as sch 
p = np.array(df_norm) 
z = linkage(df_norm, method="complete",metric="euclidean")
plt.figure(figsize=(15, 5))
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('Index')
plt.ylabel('Distance')
sch.dendrogram(
    z,
    leaf_rotation=0.,  
    leaf_font_size=8.,  
)
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering 
```

```python
h_complete = AgglomerativeClustering(n_clusters=5, linkage='complete',affinity = "euclidean").fit(df_norm)
cluster_labels=pd.Series(h_complete.labels_)
cluster_labels
air['clust']=cluster_labels 
air
```

```python
air.iloc[:,1:].groupby(air.clust).mean()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
air_subset = pd.DataFrame(scaler.fit_transform(air.iloc[:,1:]))
air_subset 
```

```python
from scipy.cluster.hierarchy import linkage 
import scipy.cluster.hierarchy as sch  
p = np.array(df_norm) 
z = linkage(air_subset, method="complete",metric="euclidean")
plt.figure(figsize=(20,10))
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('Index')
plt.ylabel('Distance')
sch.dendrogram(
    z,
    leaf_rotation=0.,  
    leaf_font_size=8., 
)
plt.show()
```
