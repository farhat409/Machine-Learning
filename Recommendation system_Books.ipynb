```python
#UBCF
import pandas as pd
import numpy as np
```

```python
books_df=pd.read_csv('book.csv')
books_df
```

```python
books_df[0:5]
```

```python
len(books_df['Book.Title'].unique())
```

```python
user_books_df=books_df.pivot_table(index='User.ID',columns='Book.Title',values='Book.Rating').reset_index(drop=True)
user_books_df
```

```python
user_books_df.index = books_df['User.ID'].unique()
```

```python
user_books_df
```

```python
#Impute those NaNs with 0 values
user_books_df.fillna(0, inplace=True)
user_books_df
```

```python
#Calculating Cosine Similarity between Users
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine, correlation
```

```python
user_sim = 1 - pairwise_distances( user_books_df.values,metric='cosine')
user_sim
```

```python
#Store the results in a dataframe
user_sim_df = pd.DataFrame(user_sim)
```

```python
user_sim_df
```

```python
#Set the index and column names to user ids 
user_sim_df.index =  books_df['User.ID'].unique()
user_sim_df.columns =  books_df['User.ID'].unique()
```

```python
user_sim_df.iloc[0:5, 0:5]
```

```python
np.fill_diagonal(user_sim, 0)
user_sim_df.iloc[0:5, 0:5]
```

```python
#Most Similar Users
user_sim_df.idxmax(axis=1)[0:5]
```

```python
Clara_user_rating = user_books_df['Decision in Normandy'] 
```

```python
Clara_user_rating
```

```python
#Finding the correlation with different movies
similar_to_Clara = user_books_df.corrwith(Clara_user_rating) 
```

```python
similar_to_Clara.head(10) 
```

```python
corr_Clara = pd.DataFrame(similar_to_Clara, columns=['Correlation'])
corr_Clara.dropna(inplace=True)
corr_Clara.head()
```

```python
corr_Clara[corr_Clara['Correlation'] > 0].sort_values(by='Correlation', ascending=False).head(5)
```
